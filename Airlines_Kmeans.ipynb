```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("/content/EastWestAirlines1.csv")
```

```python
df
```

```python
df1=df.drop(['ID#'],axis=1)
df1
```

```python
df1.info()
```

```python
ms=MinMaxScaler()
```

```python
df2=pd.DataFrame(ms.fit_transform(df1),columns=df1.columns)
```

```python
df2
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i)
  kmeans.fit(df2)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
sns.lineplot(x=range(1,11),y=wcss)
plt.title("Elbow graph")
plt.xlabel("No of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(df2)
```

```python
predict=kmeans.predict(df2)
```

```python
predict
```

```python
df1['clusterid']=predict#kmeans.labels_
```

```python
df1
```

```python
kmeans.cluster_centers_
```

```python
df1.groupby("clusterid").mean().reset_index()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df1['clusterid'],df1['Balance'],c=kmeans.labels_)
```
